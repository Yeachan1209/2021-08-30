# 코로나 이후 일상 활동 변화

# 데이터 시각화

```python
df
```

```python
df = df.set_index("항목").copy()
```

```python
df = df.drop("사례수", axis=1).copy()
```

```python
labels = ["감소", "증가", "변화 없음"]
```

```python
dex = len(df)
```

## 파이 차트

```python
dx = df.index
```

```python
# 서브플롯, 그래프의 사이즈 설정
fig, ax = plt.subplots(6, 3, figsize=(20, 40))

x = 0
y = 0

# 색깔 설정
colors = sns.color_palette('Set2')

wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

# 생성된 서브플롯에 pie chart 생성
for i in range(dex) :
  df.loc[dx[i]].plot(kind="pie", autopct="%1.1f%%", title=dx[i],
                     fontsize=15, ax=ax[x,y], shadow=True, colors=colors,
                     wedgeprops=wedgeprops, ylabel='')
  y += 1
  if y >= 3 :
    y = 0
    x += 1
  if x >= 6 :
    break  
plt.show() 
```

# 양방향 그래프

```python
index = np.arange(len(df))
```

```python
index
```

```python
# 그래프 크기 설정
plt.figure(figsize=(10, 10))
# 그래프 생성
plt.barh(index, df["증가"], color="pink", label="증가", edgecolor='Black', linewidth=3)
plt.barh(index, -df["감소"], color="green", label="감소", edgecolor='Black', linewidth=3)
# 범례 설정
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0., fontsize=15)
# 제목 이름, 폰트사이즈 설정
plt.title("코로나19 이후 국민의 일상활동 변화", fontsize=20)
# y축에 레이블 표시
plt.yticks(index, df.index, fontsize=15)
plt.grid(True)
plt.show()
```

## 스택 바 그래프

```python
# 색깔 설정
colors = sns.color_palette('Pastel1')

# 그래프 설정
ax = df.plot(kind='barh', stacked=True, fontsize=15, figsize=(15, 10),
             edgecolor='black', color=colors, linewidth=3)
plt.title("코로나19 이후 국민의 일상활동 변화", fontsize=20)
# 그래프 속 수치 추가
for x in ax.patches:
    left, bottom, width, height = x.get_bbox().bounds
    ax.annotate("%.1f"%(width), xy=(left+width/2, bottom+height/2), ha='center', va='center')
# 범례 설정
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0., fontsize=15)
# 그래프를 담고있는 박스를 없앰
plt.box(False)
plt.show()
```
